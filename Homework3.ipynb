```
import cv2
```

```
from google.colab.patches import cv2_imshow
```

```
faceCascades = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```
def face_detect(img):
   gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
   faces = faceCascades.detectMultiScale(gray,1.05, 15)
   print(faces)
   for (x,y,w,h) in faces:
     cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
     cv2.putText(img,"Face",(x,y-4),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,0,0),3)
   return img    
```

```
image = cv2.imread("AirconFixer.JPG")
image = face_detect(image)
cv2_imshow(image)
```
